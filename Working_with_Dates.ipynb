# 1. Parse the dates

```python
import pandas as pd

ibm = pd.read_csv("C:/Users/Alessandro Alvarado/Documents/OTOO 3335 Python/Data/IBM.csv", index_col = 'Date', parse_dates = True)

print(ibm.head())
```

# 2. Represent the quarterly means

```python
# This will represent the quarterly means of each column
ibm_quarter = ibm.resample('Q').mean()

print(ibm_quarter.head())
```

# 3. Plot the closing price versus the quarter designation 

```python
import matplotlib.pyplot as plt

#This will plot closing price (y-axis) versus the quarter designation (x-axis)
ibm_quarter['Close'].plot(ylabel = 'Closing Price')

plt.show()
```

# 4. Represent the weekly totals

```python
# This will represent the weekly totals of each column
ibm_weekly = ibm.resample('W').sum()

print(ibm_weekly.head())
```

# 5.  Display the 3 smallest and 3 largest volume weeks

```python
# This displays the 3 smallest volume weeks
small_volumes = ibm_weekly.nsmallest(3, 'Volume')
print(small_volumes)

# This displays the 3 largest volume weeks
large_volumes = ibm_weekly.nlargest(3, 'Volume')
print(large_volumes)
```

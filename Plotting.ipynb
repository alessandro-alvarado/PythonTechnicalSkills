# Dataset 1 â€“ Applewood_2011.csv

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the file
apple = pd.read_csv("C:/Users/Alessandro Alvarado/Documents/OTOO 3335 Python/Data/Applewood_2011.csv")
```

```python
# Turn the data into pair plot
a_plots = sns.pairplot(apple)
```

```python
# Turn the data into a bar chart
a_bar = sns.countplot(x = 'Location', hue = 'Location', data = apple, palette = 'Set2')
plt.show()
```

```python
# Turn the data into box plot
a_box = sns.boxplot(x = 'Previous', y = 'Profit', hue = 'Previous', data = apple, palette = 'Set2' )
plt.show()
```

```python
# Turn the data into scatter plots
a_scatter = sns.lmplot(x = 'Age', y = 'Profit', data = apple,
                       fit_reg = False,
                       col = 'Previous', col_wrap = 3)
```

# Dataset 3 â€“ Boston_housing.csv

```python
# Import the file
boston_housing = pd.read_csv("C:/Users/Alessandro Alvarado/Documents/OTOO 3335 Python/Data/Boston_housing.csv", sep = ";" )
```

```python
# Multiple scatter plots with lmplot and col
b_housing_lmplot = sns.lmplot(x = 'LSTAT', y = 'MEDV', col = 'CHAS', fit_reg = False, data = boston_housing)
```

```python
# Box plots
b_housing_box = sns.boxplot(x = 'CHAS', y = 'MEDV', data = boston_housing, palette = 'Set2' )

# Adding titles
b_housing_box.set_title('Median Value by Charles River Proximity')
b_housing_box.set_xlabel('Near the Charles River?')
b_housing_box.set_ylabel('Median Home Value')
```

```python
# Add custom axis labels and change other aspects of the plots.
b_housing_scatter = sns.lmplot(x = 'RM', y = 'MEDV',
                height = 4, fit_reg = False,  data = boston_housing)
b_housing_scatter.set_axis_labels('Number of Rooms', 'Median Home Value (in thousands)')
b_housing_scatter.set_titles(title = 'Number of Rooms versus Median Value')
```
